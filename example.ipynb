```python
from radar_process.fft_3d import RadarData
import torch

pre_process = RadarData((2, 4, 255, 128), (128, 128, 128))
```

```python
from scipy.io import loadmat
import numpy as np
from einops import rearrange

mat = loadmat(r"./example_data/2019_04_30_pbms002_000000.mat", squeeze_me=True)

data = np.asarray(mat["adc_data"])
data = torch.from_numpy(data)
data = rearrange(
    data, "sample (rx tx) chirp -> 1 rx tx chirp sample", rx=2, tx=4, chirp=255, sample=128
)

result = pre_process(data)
```
