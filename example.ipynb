```python
from pytorch_3dfft import RadarData
import torch

pre_process = RadarData((2, 4, 255, 128), (128, 128, 128))
```

```python
from scipy.io import loadmat
import numpy as np
from einops import rearrange

# 数据来源：https://github.com/Xiangyu-Gao/Raw_ADC_radar_dataset_for_automotive_object_detection
mat = loadmat(r"./example_data/2019_04_30_pbms002_000000.mat", squeeze_me=True)

data = np.asarray(mat["adc_data"])
data = torch.from_numpy(data)
data = rearrange(
    data, "sample (rx tx) chirp -> 1 rx tx chirp sample", rx=2, tx=4, chirp=255, sample=128
)

result = pre_process(data)
```

```python
from pytorch_3dfft import plot

plot(result[0], 0)
```

```python
from pytorch_3dfft.utils.CFAR import CFAR_OSCA_2D

cfar = CFAR_OSCA_2D(3, (40, 20), 1e-9)
pre_process = RadarData((2, 4, 255, 128), (128, 128, 128), cfar_module=cfar)
result = pre_process(data)
plot(result[0], 0)
```

```python
import plotly.graph_objects as go

plot_result = plot(result[0], 0)
plot_result.add_trace(
    go.Scatter(
        x=[40],
        y=[40],
    )
)
plot_result.show()
```
